```python
from faker import Faker
import random

fake = Faker()
country_ids = []
olympics_ids = []
players_ids = []
events_ids = []
result_ids = []
sql_result = []

def create_countries(rows_num):
    for _ in range(rows_num):
        name = fake.name()
        country_id = name[:3].upper()
        area_sqkm = fake.random_int()
        population = fake.random_int()
        country_ids.append(country_id)
        sql_string = f"insert into Countries values('{name}', '{country_id}', '{area_sqkm}', '{population}'')\n"
        sql_result.append(sql_string)
       
def create_olympics(rows_num):
    for _ in range(rows_num):
        country_id = random.choice(country_ids)
        city = fake.city()
        year = fake.unique.random_int(1896, 2021, 4)
        olympic_id = city[:3].upper() + str(year)
        startdate = fake.date_this_century()
        enddate = fake.date_this_century()
        olympics_ids.append(olympic_id)
        sql_string = f"insert into Olympics values('{country_id}', '{city}', '{year}', '{olympic_id}', '{startdate}', '{enddate}')\n"
        sql_result.append(sql_string)
        
def create_players(rows_num):
    for _ in range(rows_num):
        name = fake.name()
        player_id = fake.unique.word()
        country_id = random.choice(country_ids)
        birthdate = fake.date_of_birth()
        players_ids.append(player_id)
        sql_string = f"insert into Players values('{name}', '{player_id}', '{country_id}', '{birthdate}')\n"
        sql_result.append(sql_string)
        
def create_events(rows_num):
   for i in range(rows_num):
        event_id = 'E' + str(i)
        name = fake.word()
        eventtype = random.choice(['ATH', 'SWI'])
        olympic_id = random.choice(olympics_ids)
        is_team_event = fake.random_int(0, 1)
        if is_team_event == 1:
            num_players_in_team = fake.random_int(2, 4)
        else:
            num_players_in_team = -1
        result_noted_in = random.choice(['points', 'seconds', 'meters'])
        events_ids.append(event_id)
        sql_string = f"insert into Events values('{event_id}', '{name}', '{eventtype}', '{olympic_id}', '{is_team_event}', '{num_players_in_team}', '{result_noted_in}')\n"
        sql_result.append(sql_string)

def create_results(rows_num):
    for _ in range(rows_num):
        event_id = random.choice(events_ids)
        player_id = random.choice(players_ids)
        medal = random.choice(['GOLD', 'SILVER', 'BRONZE'])
        result = fake.pyfloat()
        sql_string = f"insert into Results values('{event_id}', '{player_id}', '{medal}', '{result}')\n"
        sql_result.append(sql_string)
        
countries_rows = 6
olympics_rows = 6
players_rows = 50
events_rows = 30
results_rows = 30

create_countries(countries_rows)
create_olympics(olympics_rows)
create_players(players_rows)
create_events(events_rows)
create_results(results_rows)

sql_file=open("insert_data.sql",mode="w",encoding="utf-8")
for sql_string in sql_result:
    sql_file.write(sql_string)
sql_file.close()
```
